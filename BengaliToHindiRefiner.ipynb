```python
def naive_bengali_sentence_split(text):
    import re
    # Split by Bengali full stop (।) and remove extra whitespace
    sentences = [s.strip() for s in re.split(r'।|\n', text) if s.strip()]
    return sentences
```

```python
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM
from transformers import M2M100Tokenizer, M2M100ForConditionalGeneration
from nltk.tokenize import sent_tokenize
```

```python
model_name='facebook/m2m100_418M'
tokenizer=M2M100Tokenizer.from_pretrained(model_name)
model=M2M100ForConditionalGeneration.from_pretrained(model_name)
```

```python
def translate_bengali_to_hindi(bn_txt):
  sentences=naive_bengali_sentence_split(bn_txt)
  translations=[]
  for sentence in sentences:
    tokenizer.src_lang='bn'
    encoded_txt=tokenizer(sentence,return_tensors='pt')
    output=model.generate(**encoded_txt, forced_bos_token_id=tokenizer.get_lang_id('hi'),max_length=512)
    hi_txt=tokenizer.decode(output[0],skip_special_tokens=True)
    translations.append(hi_txt)
  hi_txt=' '.join(translations)
  return hi_txt
```

```python
print(text)
hindi=translate_bengali_to_hindi(text)
print(hindi)
```
